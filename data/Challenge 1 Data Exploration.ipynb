## Correlation and Apparent Relationships

```python
num_cols = ["age", "area", "height"] 
sns.pairplot(df[num_cols], size=5)
```

```python
df_num = df[['age', 'area', 'height']]
df_num.corr()
```

```python
## Create boxplots of data
def group_boxplot(df, plot_cols, by):
    plt.style.use('bmh')
    for col in plot_cols:
        fig = plt.figure(figsize=(8,8))
        ax = fig.gca()
        df.boxplot(column = col, by = by, ax = ax)
        ax.set_title('Box plots of ' + col + ' by ' + by)
        ax.set_ylabel(col)
    return by 

## Define columns for making a conditioned histogram
plot_cols2 = ["age",
               "area",
               "height"]

group_boxplot(df, plot_cols2, "damage_grade")
```
